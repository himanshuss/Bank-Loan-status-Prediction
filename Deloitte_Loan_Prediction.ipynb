# Importing the Relevant Libraries

from pycaret.utils import enable_colab 
enable_colab()

```python
import numpy as np
import pandas as pd
from pycaret.classification import *
```

# Importing the Training Dataset

```python
raw_data = pd.read_csv('train.csv')
raw_data.head(10)
```

### Descriptive Statistics

```python
raw_data.describe(include='all')
```

### Features of Interest

```python
raw_data = raw_data.drop(['ID', 'Sub Grade', 'Payment Plan', 'Loan Title', 
                          'Accounts Delinquent','Collection Recovery Fee','Total Received Late Fee',
                         'Total Received Interest','Recoveries','Total Collection Amount'], axis=1)
raw_data.info()
```

##### Note:- There are no missing values

### Dealing with Categorical Data

```python
raw_data['Grade'] = raw_data['Grade'].map({'A':0,
                                           'B':1,
                                           'C':2,
                                           'D':3,
                                           'E':4,
                                           'F':5,
                                           'G':6})
```

```python
raw_data['Employment Duration'] = raw_data['Employment Duration'].map({'MORTGAGE':0,
                                                                       'RENT':1,
                                                                       'OWN':2})
```

```python
raw_data['Verification Status'] = raw_data['Verification Status'].map({'Not Verified':0,
                                                                       'Verified': 1,
                                                                       'Source Verified': 1})
```

```python
raw_data['Initial List Status'] = raw_data['Initial List Status'].map({'w':0,
                                                                       'f':1})
```

```python
raw_data['Application Type'] = raw_data['Application Type'].map({'INDIVIDUAL':0,
                                                                 'JOINT':1})
```

# Experimenting different models (PyCaret)

```python
categorical_variables = ['Grade', 'Employment Duration', 'Verification Status',
                         'Verification Status', 'Initial List Status', 'Application Type',
                         'Delinquency - two years', 'Inquires - six months', 'Public Record',
                         'Collection 12 months Medical','Batch Enrolled']

experiment = setup(data=raw_data, target='Loan Status', categorical_features=categorical_variables)
```

### Find the best model

```python
best_model = compare_models()
```

```python
predict_model(best_model, raw_data.tail())
```

### Save the Best Model

```python
save_model(best_model, model_name='Gradient Boosting Classifier')
```

### Loading the best Model

```python
model = load_model('Logistic Regression Model')
```

# Loading the Testing Dataset

```python
data_test = pd.read_csv('test.csv')

data_test = data_test.drop(['ID', 'Batch Enrolled', 'Sub Grade', 'Payment Plan', 'Loan Title', 'Accounts Delinquent', 'Loan Status'], axis=1)

data_test['Grade'] = data_test['Grade'].map({'A':0,
                                           'B':1,
                                           'C':2,
                                           'D':3,
                                           'E':4,
                                           'F':5,
                                           'G':6})

data_test['Employment Duration'] = data_test['Employment Duration'].map({'MORTGAGE':0,
                                                                       'RENT':1,
                                                                       'OWN':2})

data_test['Verification Status'] = data_test['Verification Status'].map({'Not Verified':0,
                                                                       'Verified': 1,
                                                                       'Source Verified': 1})

data_test['Initial List Status'] = data_test['Initial List Status'].map({'w':0,
                                                                       'f':1})

data_test['Application Type'] = data_test['Application Type'].map({'INDIVIDUAL':0,
                                                                 'JOINT':1})
```

### Predicting the Testing Dataset

```python
prediction_data = predict_model(best_model, data_test)
prediction_data.rename(columns={'Label':'Loan Status'}, inplace=True)
prediction_data
```

```python
prediction_data_Status = prediction_data['Loan Status']
prediction_data_Status
```

```python
model.predict(data_test)
```
